```python
# üìä Anime Dataset - Exploratory Data Analysis (EDA)
# Save this as: notebooks/01_data_exploration.ipynb

# Cell 1: Markdown
"""
# üìä Anime Dataset - Exploratory Data Analysis

**Objective:** Explore and understand the anime dataset through statistical analysis and visualizations.

**Contents:**
1. Data Loading
2. Basic Statistics
3. Missing Values Analysis
4. Distribution Analysis
5. Correlation Analysis
6. Key Insights
"""
```

```python
# Cell 2: Setup and Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import sys
from pathlib import Path

# Add parent directory to path
notebook_dir = Path.cwd()
project_root = notebook_dir.parent
sys.path.insert(0, str(project_root))

# Now import from project
from config import ANIME_RAW, ANIME_CLEANED

# Suppress warnings
warnings.filterwarnings('ignore')

# Set visualization style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['font.size'] = 10

print("‚úÖ Imports successful!")
print(f"üìÅ Project root: {project_root}")
```

```python
# Cell 3: Load Data
print("="*70)
print("LOADING DATA")
print("="*70)

# Load anime data
anime_df = pd.read_csv(ANIME_RAW)

print(f"\n‚úÖ Dataset loaded successfully!")
print(f"üìä Shape: {anime_df.shape}")
print(f"üìã Columns: {len(anime_df.columns)}")

# Display first few rows
anime_df.head()
```

```python
# Cell 4: Basic Information
print("\n" + "="*70)
print("DATASET OVERVIEW")
print("="*70)

print(f"\nüì∫ Total Anime: {len(anime_df):,}")
print(f"üìã Total Features: {len(anime_df.columns)}")
print(f"\nüìù Column Names:")
for i, col in enumerate(anime_df.columns, 1):
    print(f"  {i:2d}. {col}")

print("\nüìä Data Types:")
anime_df.info()
```

```python
# Cell 5: Descriptive Statistics
print("\n" + "="*70)
print("DESCRIPTIVE STATISTICS")
print("="*70)

anime_df.describe()
```

```python
# Cell 6: Missing Values Analysis
print("\n" + "="*70)
print("MISSING VALUES ANALYSIS")
print("="*70)

# Calculate missing values
missing = anime_df.isnull().sum()
missing_pct = (missing / len(anime_df) * 100).round(2)

missing_df = pd.DataFrame({
    'Column': missing.index,
    'Missing Count': missing.values,
    'Percentage': missing_pct.values
})

# Filter and sort
missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values(
    'Missing Count', ascending=False
)

print("\nüìä Columns with Missing Values:")
print(missing_df.to_string(index=False))

# Visualize
if len(missing_df) > 0:
    fig, ax = plt.subplots(figsize=(12, 6))
    ax.barh(missing_df['Column'], missing_df['Percentage'], color='coral')
    ax.set_xlabel('Missing Percentage (%)')
    ax.set_title('Missing Values by Feature', fontsize=14, fontweight='bold')
    ax.grid(alpha=0.3, axis='x')
    plt.tight_layout()
    plt.show()
else:
    print("\n‚úÖ No missing values detected!")
```

```python
# Cell 7: Score Distribution Analysis
print("\n" + "="*70)
print("SCORE DISTRIBUTION ANALYSIS")
print("="*70)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Histogram with KDE
axes[0].hist(anime_df['Score'].dropna(), bins=50, edgecolor='black', alpha=0.7)
axes[0].axvline(anime_df['Score'].mean(), color='red', linestyle='--', 
                linewidth=2, label=f"Mean: {anime_df['Score'].mean():.2f}")
axes[0].axvline(anime_df['Score'].median(), color='green', linestyle='--', 
                linewidth=2, label=f"Median: {anime_df['Score'].median():.2f}")
axes[0].set_xlabel('Score')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Score Distribution', fontsize=14, fontweight='bold')
axes[0].legend()
axes[0].grid(alpha=0.3)

# Box plot
axes[1].boxplot(anime_df['Score'].dropna(), vert=True)
axes[1].set_ylabel('Score')
axes[1].set_title('Score Box Plot', fontsize=14, fontweight='bold')
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.show()

# Statistics
print(f"\nüìä Score Statistics:")
print(f"  Mean:     {anime_df['Score'].mean():.2f}")
print(f"  Median:   {anime_df['Score'].median():.2f}")
print(f"  Std Dev:  {anime_df['Score'].std():.2f}")
print(f"  Min:      {anime_df['Score'].min():.2f}")
print(f"  Max:      {anime_df['Score'].max():.2f}")
print(f"  Q1 (25%): {anime_df['Score'].quantile(0.25):.2f}")
print(f"  Q3 (75%): {anime_df['Score'].quantile(0.75):.2f}")
```

```python
# Cell 8: Genre Analysis
print("\n" + "="*70)
print("GENRE ANALYSIS")
print("="*70)

# Extract all genres
all_genres = anime_df['Genres'].str.split(',').explode().str.strip()
genre_counts = all_genres.value_counts()

print(f"\nüìä Total Unique Genres: {genre_counts.nunique()}")
print(f"\nüèÜ Top 10 Most Popular Genres:")
print(genre_counts.head(10))

# Visualize top 15
fig, ax = plt.subplots(figsize=(12, 8))
genre_counts.head(15).plot(kind='barh', ax=ax, color='skyblue')
ax.set_xlabel('Count')
ax.set_ylabel('Genre')
ax.set_title('Top 15 Anime Genres', fontsize=16, fontweight='bold')
ax.invert_yaxis()
ax.grid(alpha=0.3, axis='x')
plt.tight_layout()
plt.show()
```

```python
# Cell 9: Type Distribution
print("\n" + "="*70)
print("TYPE DISTRIBUTION")
print("="*70)

type_counts = anime_df['Type'].value_counts()
print("\nüìä Anime Types:")
print(type_counts)

# Pie chart
fig, ax = plt.subplots(figsize=(10, 8))
colors = plt.cm.Set3(range(len(type_counts)))
ax.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%', 
       startangle=90, colors=colors)
ax.set_title('Anime Type Distribution', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Cell 10: Top Anime Analysis
print("\n" + "="*70)
print("TOP ANIME BY SCORE")
print("="*70)

top_anime = anime_df.nlargest(20, 'Score')[
    ['Name', 'Score', 'Genres', 'Type', 'Episodes', 'Members']
]

print("\nüèÜ Top 20 Highest Rated Anime:")
print(top_anime.to_string(index=False))

# Visualize
fig, ax = plt.subplots(figsize=(12, 10))
y_pos = np.arange(len(top_anime))
ax.barh(y_pos, top_anime['Score'], color='coral')
ax.set_yticks(y_pos)
ax.set_yticklabels(top_anime['Name'])
ax.set_xlabel('Score')
ax.set_title('Top 20 Anime by Score', fontsize=16, fontweight='bold')
ax.invert_yaxis()
ax.grid(alpha=0.3, axis='x')
plt.tight_layout()
plt.show()
```

```python
# Cell 11: Correlation Analysis
print("\n" + "="*70)
print("CORRELATION ANALYSIS")
print("="*70)

# Select numeric columns
numeric_cols = ['Score', 'Episodes', 'Members', 'Favorites']
# Only use columns that exist
numeric_cols = [col for col in numeric_cols if col in anime_df.columns]

if len(numeric_cols) >= 2:
    numeric_data = anime_df[numeric_cols].dropna()
    
    # Correlation matrix
    corr_matrix = numeric_data.corr()
    
    print("\nüìä Correlation Matrix:")
    print(corr_matrix)
    
    # Heatmap
    fig, ax = plt.subplots(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm',
                center=0, square=True, linewidths=1, ax=ax)
    ax.set_title('Feature Correlation Heatmap', fontsize=16, fontweight='bold')
    plt.tight_layout()
    plt.show()
    
    # Key insights
    print("\nüí° Key Correlations:")
    for i in range(len(corr_matrix)):
        for j in range(i+1, len(corr_matrix)):
            corr_val = corr_matrix.iloc[i, j]
            if abs(corr_val) > 0.3:  # Strong correlation
                print(f"  ‚Ä¢ {corr_matrix.index[i]} ‚Üî {corr_matrix.columns[j]}: {corr_val:.2f}")
```

```python
# Cell 12: Popularity vs Score
print("\n" + "="*70)
print("POPULARITY VS QUALITY ANALYSIS")
print("="*70)

if 'Members' in anime_df.columns:
    fig, ax = plt.subplots(figsize=(12, 6))
    
    # Scatter plot
    ax.scatter(anime_df['Members'], anime_df['Score'], 
               alpha=0.3, s=20, c=anime_df['Score'], cmap='viridis')
    
    ax.set_xlabel('Members (Popularity)', fontsize=12)
    ax.set_ylabel('Score (Quality)', fontsize=12)
    ax.set_title('Anime Popularity vs Quality', fontsize=14, fontweight='bold')
    ax.set_xscale('log')
    ax.grid(alpha=0.3)
    
    # Add colorbar
    cbar = plt.colorbar(ax.collections[0], ax=ax)
    cbar.set_label('Score', rotation=270, labelpad=15)
    
    plt.tight_layout()
    plt.show()
    
    # Calculate correlation
    corr = anime_df[['Members', 'Score']].corr().iloc[0, 1]
    print(f"\nüìä Correlation between Popularity and Score: {corr:.3f}")
    
    if corr > 0.3:
        print("  ‚úÖ Positive correlation: Popular anime tend to have higher scores")
    elif corr < -0.3:
        print("  ‚ö†Ô∏è Negative correlation: Popular anime tend to have lower scores")
    else:
        print("  ‚ÑπÔ∏è Weak correlation: Popularity and score are somewhat independent")
```

```python
# Cell 13: Episodes Distribution
print("\n" + "="*70)
print("EPISODES DISTRIBUTION")
print("="*70)

if 'Episodes' in anime_df.columns:
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))
    
    # All episodes
    axes[0].hist(anime_df['Episodes'].dropna(), bins=50, 
                 edgecolor='black', alpha=0.7)
    axes[0].set_xlabel('Episodes')
    axes[0].set_ylabel('Count')
    axes[0].set_title('Episode Distribution (All)', fontsize=12, fontweight='bold')
    axes[0].grid(alpha=0.3)
    
    # Filtered (‚â§50 episodes)
    filtered = anime_df[anime_df['Episodes'] <= 50]['Episodes']
    axes[1].hist(filtered.dropna(), bins=50, 
                 edgecolor='black', alpha=0.7, color='orange')
    axes[1].set_xlabel('Episodes')
    axes[1].set_ylabel('Count')
    axes[1].set_title('Episode Distribution (‚â§50 episodes)', 
                      fontsize=12, fontweight='bold')
    axes[1].grid(alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    print(f"\nüìä Episode Statistics:")
    print(f"  Mean:   {anime_df['Episodes'].mean():.1f}")
    print(f"  Median: {anime_df['Episodes'].median():.0f}")
    print(f"  Max:    {anime_df['Episodes'].max():.0f}")
```

```python
# Cell 14: Summary and Key Insights
print("\n" + "="*70)
print("EDA SUMMARY & KEY INSIGHTS")
print("="*70)

print(f"\nüìä Dataset Overview:")
print(f"  ‚Ä¢ Total Anime: {len(anime_df):,}")
print(f"  ‚Ä¢ Features: {len(anime_df.columns)}")
print(f"  ‚Ä¢ Average Score: {anime_df['Score'].mean():.2f}/10")
print(f"  ‚Ä¢ Total Genres: {all_genres.nunique()}")
print(f"  ‚Ä¢ Most Common Type: {type_counts.index[0]} ({type_counts.values[0]:,} anime)")

print(f"\nüí° Key Insights:")
print(f"  1. Score Distribution:")
print(f"     ‚Ä¢ Scores are slightly left-skewed")
print(f"     ‚Ä¢ Most anime have scores between 6.0-8.0")
print(f"     ‚Ä¢ Very few anime have extremely low (<4.0) or high (>9.0) scores")

print(f"\n  2. Genre Trends:")
top_3_genres = genre_counts.head(3)
print(f"     ‚Ä¢ Top 3 genres: {', '.join(top_3_genres.index.tolist())}")
print(f"     ‚Ä¢ These account for {(top_3_genres.sum() / genre_counts.sum() * 100):.1f}% of all genre tags")

print(f"\n  3. Content Types:")
print(f"     ‚Ä¢ {type_counts.index[0]} dominates with {type_counts.values[0]:,} entries")
print(f"     ‚Ä¢ Distribution: {', '.join([f'{t} ({c})' for t, c in zip(type_counts.index[:3], type_counts.values[:3])])}")

if 'Members' in anime_df.columns and corr > 0:
    print(f"\n  4. Quality vs Popularity:")
    print(f"     ‚Ä¢ Positive correlation ({corr:.2f}) detected")
    print(f"     ‚Ä¢ Popular anime generally have higher ratings")
    print(f"     ‚Ä¢ Suggests quality content gains more viewers")

print(f"\n‚úÖ EDA Complete! Ready for data cleaning and modeling.")
```
